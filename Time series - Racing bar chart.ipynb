```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
#Ignore warnings created while preparing charts

import warnings
warnings.filterwarnings("ignore")
```

```python
countries = pd.read_csv("noc_regions.csv")
medals = pd.read_csv("athlete_events.csv")
```

```python
medals.info()
```

```python
medals.head()
```

```python
countries.info()
```

```python
#Merging dataframes ot get countrinames against medal counts
df = pd.merge(medals, countries, on = 'NOC')
df.info()
```

```python
#Selecting only the required columns
agg_df = df[['Year','Medal','region']].copy()
agg_df.info()
```

```python
#Replacing no medals as 0
agg_df['Medal'].fillna(0, inplace = True)
agg_df.info()
```

```python
#Identifying the records with no region information
agg_df[agg_df['region'].isna()]
```

```python
#Removing records having no information on region
agg_df.dropna(axis = 0, inplace = True)
agg_df.info()
```

```python
#Identifying overall medal counts
agg_df.Medal.value_counts()
```

```python
#Removing records with no medals
medals_df = agg_df[agg_df['Medal'] != 0]
medals_df
```

```python
#Creating pivot table with overall medal counts 
medals_df = medals_df.pivot_table(columns = 'region', index = 'Year', values='Medal', aggfunc = 'count')
medals_df
```

```python
#Filling records with no medals as 0
medals_df.fillna(0, inplace = True)
```
